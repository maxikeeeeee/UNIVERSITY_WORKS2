```python
import numpy as np
from math import factorial
import matplotlib.pyplot as plt

x=[0.101, 0.106, 0.111, 0.116, 0.121, 0.126]
y=[1.2618 ,1.2764, 1.2912, 1.3061, 1.3213, 1.3366]
h = x[1] - x[0]
x1=0.1
x2=0.9
q=(x1 - x[0])/h
q1 = (x2-x[-1])/h
print('Крок таблиці h =', h)
print('Для першої формули Ньютона q =', q ,'; Для другої формули Ньютона q =', q1)
def n(y,j):
    mas=[]
    for i in range(len(y)):
        mas.append(y[i] - y[i-1])
    mas.pop(0)
    if j == 1:
        return mas
    else:
        j-=1
        return n(mas, j)
#Перша формула Ньютона
newton1 = (y[0]+q*n(y,1)[0]+q*(q-1)*n(y,2)[0]/factorial(2))+(q*(q-1)*(q-2)*n(y,3)[0]/factorial(3))+(q*(q-1)*(q-2)*(q-3)*n(y,4)[0]/factorial(4))+(q*(q-1)*(q-2)*(q-3)*(q-4)*n(y,5)[0]/factorial(5))

print ('Значення функції в точці х1=', x1, 'Перша формула Ньютона', round(newton1,5))
#Друга формула Ньютона
newton2 = (y[5] + q1*n(y,1)[4]+q1*(q1+1)*n(y,2)[3]/factorial(2))+(q1*(q1+1)*(q1+2)*n(y,3)[2]/factorial(3))+(q1*(q1+1)*(q1+2)*(q1+3)*n(y,4)[1]/factorial(4))+(q1*(q1+1)*(q1+2)*(q1+3)*(q1+4)*n(y,4)[1]/factorial(5))

print ('Значення функції в точці х2=', x2, 'Друга формула Ньютона', round(newton2,5))
```

```python
x_1=np.linspace(np.min(x), np.max(x))
y_1=np.linspace(np.min(y), np.max(y))
plt.plot(x,y, 'ro', x_1, y_1)
plt.title('Графік інтерполяційної функції')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
