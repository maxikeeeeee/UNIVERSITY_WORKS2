```python
# Імпортуємо бібліотеки
import numpy as np
import matplotlib.pyplot as plt

# Задаємо точки
xi = np.array([-1, 0, 2, 3], dtype=float)
yi = np.array([-16, -7, -1, 20], dtype=float)

# Функція Лагранжа
def lagranzha(x, y, t):
    z = 0
    for j in range(len(y)):
        p1 = 1
        p2 = 1
        for i in range(len(x)):
            if i == j:
                p1 *= 1
                p2 *= 1
            else:
                p1 *= (t - x[i])
                p2 *= (x[j] - x[i])
        z += y[j] * (p1 / p2)
    return z

x_new = np.linspace(np.min(xi), np.max(xi), 100)
y_new = [lagranzha(xi, yi, i) for i in x_new]

# Виводимо графік
plt.plot(x_new, y_new)
plt.plot(xi, yi, 'o', x_new, y_new)
plt.title('Графік функції Лагранжа')
plt.grid(True)
plt.show()
```

```python
# Задамо функцію
def f(x):
    return 2*x**3 - 4*x**2 + 3*x - 7

# Задаємо точки для інтерполяції
x_custom = np.array([-0.5, 0.5, 1.5, 2.5], dtype=float)
# Знаходимо значення функції в точках
y_custom = [f(i) for i in x_custom]

# Виводимо результат
print("Значення функції в точках:")
for i in range(len(x_custom)):
    print(f"f({x_custom[i]: }) = L({x_custom[i]: }) = {y_custom[i]: }")
```

```python
# Імпортуємо lagrange
from scipy.interpolate import lagrange

#Перевірка 
f = lagrange(xi, yi)
plt.plot(x_new, f(x_new), 'b', xi, yi, 'ro')
plt.title('Графік функції Лагранжа (SciPy)')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
