```python
import numpy as np

def f1(x):
    return 0.5 - np.cos(x - 1)

def f2(y):
    return 3 + np.cos(y)
```

```python
import matplotlib.pyplot as plt

x1 = np.arange(1, 6, 0.1)
y1 = [f1(i) for i in x1]
plt.plot(x1, y1, 'r')

y2 = np.arange(-2, 5, 0.1)
x2 = [f2(i) for i in y2]
plt.plot(x2, y2, 'b')

plt.grid()
plt.xlabel(xlabel='x')
plt.ylabel(ylabel='y')
```

```python
# Range: 3 < x < 3.5; 1 < y < 1.5
xa = 3; xb = 3.5
ya = 1; yb = 1.5
```

```python
from scipy.misc import derivative

# d(f1)/dx = derivative(f1); d(f2)/dx = 0
# d(f1)/dy = 0; d(f2)/dy = derivative(f2)

der1 = abs(derivative(f1, xb))
der2 = abs(derivative(f2, yb))

print(f"Derivative 1: {der1:.4g} < 1")
print(f"Derivative 2: {der2:.4g} < 1")
```

```python
xn = (xa + xb) / 2
yn = (ya + yb) / 2

while True:
    yn1 = f1(xn)
    xn1 = f2(yn)
    if abs(xn1 - xn) < 0.0001 and abs(yn1 - yn) < 0.0001:
        break
    yn = yn1
    xn = xn1

print(f"Iteration Solution: ({xn:.4f}, {yn:.4f})")
```

```python
from scipy.optimize import fsolve

def f3(p):
    x, y = p
    return np.cos(x - 1) - y - 0.5, x - np.cos(y) - 3

s = fsolve(f3, [1.,1.])
print(f"SciPy Solution: {s}")
```
